```python
import os
import pandas as pd

df = pd.DataFrame()

# iterate through each file in the directory
for file in os.listdir('.'):
    if file.endswith(".txt"):
        # add file contents to running df
        df = df.append(pd.read_csv(file, sep='\t'))
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
smi_df = df["SMILES"]
```

```python
smi_df = smi_df.dropna().reset_index(drop=True)
smi_df
```

```python
smi_df = smi_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
smi_df.to_csv("surechembl_smiles.csv", index=False)
```
