```python
import pandas as pd
```

```python
df_preds = pd.read_csv('test_preds.csv')
df_targets = pd.read_csv('test_targets.csv')

# sort df_preds by hcv
df_preds = df_preds.sort_values(by=['hcv'], ascending=False)

# set cid to index
df_preds = df_preds.set_index('cid')
df_preds = df_preds.T

df_targets = df_targets.set_index('cid')
df_targets = df_targets.T

df_preds.to_csv('test_preds_T.csv')
df_targets.to_csv('test_targets_T.csv')
```
