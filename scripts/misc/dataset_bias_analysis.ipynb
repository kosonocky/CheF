```python
import pandas as pd
from ast import literal_eval
from collections import Counter

df = pd.read_csv("../../results/pre_final_files/schembl_summs_v2_fixed_errors.csv")
```

```python
df["summarizations"] = df["summarizations"].apply(literal_eval)
```

```python
all_summs = Counter()
for _set in df["summarizations"].tolist():
    for summ in _set:
        all_summs[summ] += 1
```

```python
# create bar plot with counter frequencies on y axis and label on x axis. In order of most frequent to least frequent. ONLY THE TOP 20

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 10))

# get top 20
top_25 = all_summs.most_common(25)

# get labels and frequencies
labels = [x[0] for x in top_25][::-1]
freqs = [x[1] for x in top_25][::-1]

# plot
plt.barh(labels, freqs)
plt.xlabel("Frequency")
plt.ylabel("Summarization")
# plt.title("Frequency of Summarizations in SchemBL Summaries")
plt.savefig("summarization_freqs_top_25_with_labels.png", bbox_inches="tight")
plt.show()
```

```python
# now plot frequencies of all summarizations as a line plot (ignore labels this time)

plt.figure(figsize=(10, 10))

# get labels and frequencies
labels = [x[0] for x in all_summs.most_common()[:100]]
freqs = [x[1] for x in all_summs.most_common()[:100]]

# plot
plt.plot(freqs)
plt.xlabel("Summarization")
plt.ylabel("Frequency")
plt.savefig("summarization_freqs_top_100.png", bbox_inches="tight")
plt.show()
```
