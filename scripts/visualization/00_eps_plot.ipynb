```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pathlib import Path
```

```python
path = Path("../../results/eps_diffs/")

eps_list = []
n_clusters_list = []

# for file in path, print filename
for file in sorted(path.iterdir()):
    if "eps" in file.name:
        split_file = file.name.split("_")
        eps = float(split_file[1])
        n_clusters = int(split_file[3])
        eps_list.append(eps)
        n_clusters_list.append(n_clusters)

```

```python
df = pd.DataFrame()
df["eps"] = eps_list
df["n_clusters"] = n_clusters_list
```

```python

plt.style.use('seaborn-darkgrid')
fig = sns.lineplot(x="eps", y="n_clusters", data=df, marker='o', color='orange', linewidth=2, alpha=0.7 )

# get index of 0.340
idx = df.index[df['eps'] == 0.340].tolist()[0]
# plot eps and n_clusters at index in black
plt.plot(df['eps'][idx], df['n_clusters'][idx], 'ko', marker="*", markersize=10, alpha=0.7)

fig.set(xlabel="Epsilon")
fig.set(ylabel="Number of clusters")

# make x range go from 0 to 1
fig.set_xlim(0,1)
fig.figure.savefig("eps_by_n_clusters.png", dpi=300)
```
