```python
# for file in ../results
import os
from pathlib import Path
for file in os.listdir("../results/eps_diffs"):
    if file.endswith("_clusters.txt"):
        print(file)
        with open(f"../results/eps_diffs/{file}", "r") as f:
            clusters = f.read().splitlines()
        # sort by length of cluster
        clusters = sorted(clusters, key = lambda x: len(x.split(", ")), reverse = True)
        # save to new file
        with open(f"../results/eps_diffs/{file}", "w") as f:
            for item in clusters:
                f.write(f"{item}\n")
    


# # save
# with open("../results/eps_diffs/eps_030_diff_49640_clusters_sorted.txt", "w") as f:
#     for item in clusters:
#         f.write(f"{item}\n")
```

```python
from collections import Counter

with open("../results/schembl_summs_v3_alg_cleaned_vocab.txt", "r") as f:
    labels = f.read().splitlines()

# split each label by spaces
labels = [x.split() for x in labels]

# count the number of labels
label_counts = Counter([item for sublist in labels for item in sublist])
```

```python
# print each label in label_counts sorted by count
for label, count in sorted(label_counts.items(), key = lambda x: x[1], reverse = True):
    print(f"{label}: {count}")
```
